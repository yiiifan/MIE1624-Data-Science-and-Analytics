## Data cleaning
- Mainly three parts of data preprocessing:
    - remove noise: remove HTML tags/URLs/punctuaions, replace HTML character with an ASCII equivalent
    - tokenization: seperate the string into words
    - normalization: convert each word to lowercase, replace unrecognizable characters and Stemming.

```python
!pip install contractions
!pip install inflect
!pip install -U nltk
!pip install regex
!pip install unicodedata2
!pip install beautifulsoup4
```

```python
import re, string, unicodedata
import html
import nltk
nltk.download('wordnet')
import contractions
import inflect
from bs4 import BeautifulSoup
from nltk.corpus import stopwords
from nltk.stem import LancasterStemmer, WordNetLemmatizer
from nltk.tokenize import TweetTokenizer
```

```python
def strip_html(text):
    """ Remove html tags and attributes """
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

def remove_char(text):
    """ Replace html character codes with equivalent ASCII code """
    return html.unescape(text)

def remove_url(text):
    """ Remove URLs """
    return re.sub(r"http\S+", "", text, flags=re.MULTILINE)

def characterization(text):
    """ Remove puctuations, numbers and special characters"""
    reg = re.compile(r'[^\w\s]')
    return reg.sub('',text)
```

```python
def remove_noise(text):
    text = strip_html(text)
    text = remove_char(text)  
    text = remove_url(text)      
    text = characterization(text)
    return text
```

```python
def tokenization(text):
    """ Splict text into words """
    tknzr = TweetTokenizer()
    return tknzr.tokenize(text)
```

```python
def create_stopwords():
    """ Create the list of stopwords by stop_words.txt """
    stopwords = []
    st = open('stop_words.txt','r',encoding='UTF-8').read()    
    stList = st.split('\n')
    return stList

def remove_stopwords(words, stList):
    """ Remove stop words from list of tokenized words """
    new_words = []
    for word in words:
        if word not in stList:
            new_words.append(word)
    return new_words
```

```python
def normalization(words, stList):
    words = remove_stopwords(words, stList)
    new_words = []
    stemmer = LancasterStemmer()
    lemmatizer = WordNetLemmatizer()
    for word in words:
        '''convert each word to lowercase'''
        new_word = word.lower()
        '''replace unrecognizable characters'''        
        new_word = unicodedata.normalize('NFKD', new_word).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        '''Stemming and Lemmatizing'''
        new_word = stemmer.stem(new_word)
        new_word = lemmatizer.lemmatize(new_word, pos='v')
        new_words.append(new_word)
    return new_words
```

```python
''' Preprocess the US airline dataset '''
stList = create_stopwords()
tidy_tweet = []

length = len(us_tweets)
for i in range(length):
    text = us_tweets.iloc[i]['text']
    text = remove_noise(text)
    words = tokenization(text)
    words = normalization(words, stList)
    tidy_tweet.append(' '.join(words)) 
    
us_tweets['tidy_tweets'] = pd.Series(tidy_tweet)
us_tweets[0:5]
```

```python
''' Preprocess the generic tweets dataset '''
gen_len = len(gen_tweets)
tweet = []
for i in range(gen_len):
    text = gen_tweets.iloc[i]['text']
    text = remove_noise(text)
    words = tokenization(text)
    words = normalization(words, stList)
    tweet.append(' '.join(words)) 

gen_tweets['tidy_tweets'] = pd.Series(tweet)
gen_tweets[0:5]
```

## Model Preparation

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
```

### Use tfidf package to calculate the frequency of each word as the features of the model

```python
tfidf_vectorizer = TfidfVectorizer(max_df=0.90, min_df=2)

tfidf = tfidf_vectorizer.fit_transform(gen_tweets['tidy_tweets'])
us_tfidf = tfidf_vectorizer.transform(us_tweets['tidy_tweets'])
```

### Split the generic tweets randomly into training data and test data

```python
x_train, x_test, y_train, y_test = train_test_split(tfidf, gen_tweets['class'], random_state=42, test_size=0.3)
```

## Model Implementation
### Train the model and obtain an acurracy

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
lreg = LogisticRegression()
lreg.fit(x_train, y_train)
prediction = lreg.predict(x_test)
print('Accuracy of logistic regression classifier on train set: {:.5f}'.format(lreg.score(x_train, y_train)))
print('Accuracy of logistic regression classifier on test set: {:.5f}'.format(lreg.score(x_test, y_test)))
```

### Evaluate the same model on the US airline data

```python
for i in range(length):
    senti = us_tweets.iloc[i]['sentiment']
    if senti == "positive":
        us_tweets.loc[i,'class'] = 4
    else:
        us_tweets.loc[i,'class'] = 0

print('Accuracy of logistic regression classifier on train set: {:.5f}'.format(lreg.score(us_tfidf,us_tweets["class"])))
us_tweets[0:3]
```

### 4. Suggest one way you can improve the accuracy of you models.

- Tuning hyperparameter to improve the accuracy(**Bonus**).

```python
def tune_hyper(x_train, x_test, y_train, y_test):
    result_list = []
    
    for i in np.logspace(-5,1.0,60):
        lr = LogisticRegression(penalty = 'l1', C = i,random_state = 0)
        lr.fit(x_train,y_train)
        
        y_train_pred = lr.predict(x_train)
        y_test_pred = lr.predict(x_test)
        train_accur = accuracy_score(y_train,y_train_pred)
        test_accur = accuracy_score(y_test,y_test_pred)
        result_list.append((i, train_accur, test_accur))

    result_df = pd.DataFrame(result_list, columns=["param", "train_accuracy", "test_accuracy"])
    return result_df
```

```python
# x_train, x_test, y_train, y_test= train_test_split(us_tfidf, us_tweets['sentiment'], random_state=42, test_size=0.3)
result = tune_hyper(x_train, x_test, y_train, y_test)
result.plot(x="param", y=["train_accuracy", "test_accuracy"], logx=True)
```

```python
re_xtrain, re_xtest, re_ytrain, re_ytest = train_test_split(re_tfidf, df['negative_reason'], random_state=42, test_size=0.3)
result = tune_hyper(re_xtrain, re_xtest, re_ytrain, re_ytest)
result.plot(x="param", y=["train_accuracy", "test_accuracy"], logx=True)
```
